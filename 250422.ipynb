```python
df1 = {'name': ['foo','bar', 'bae', 'fool'],'age': [1, 2, 3, 5]}
df2 = {'name': ['foo', 'bar', 'baz', 'foo'],'age': [1, 2, 7, 8],'salary': [34,23,4,5]}
```

```python
a=pd.DataFrame(df1)
```

```python
b=pd.DataFrame(df2)
```

```python
pd.merge(a,b,how='inner')
#inner joint: common data will print
#left joint: left joint will print
#outer joint: whole data will print(no common)
```

```python
pd.merge(a,b,how='left')#it is the first data
```

```python
pd.merge(a,b,how='right')#the second data that is df2
```

```python
pd.merge(a,b,how='outer') #whole data
```

```python
#concatenate
#append
df_1 = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D'],
    'math': [68,69,42,79],
    'physics': [46,99,80,64],
    'chemistry': [31,91,77,70]
})

df_2 = pd.DataFrame({
    'name': ['E', 'F', 'G', 'H'],
    'math': [66,95,83,66],
    'physics': [63,89,88,70],
    'chemistry': [100,81,73,90],
    'biology': [56,90,78,56]
})
```

```python
pd.concat([df_1, df_2],ignore_index=True)
```

```python
pd.concat([df_1, df_2], ignore_index=True, join="inner")
```
