```python
import seaborn as sns
import pandas as pd
```

```python
days=[1,2,3,4,5,6,7]
temp=[34,45,23,12,0,45,33]
sns.lineplot(days,temp)
```

```python
days=[1,2,3,4,5,6,7]
temp=[34,45,23,12,0,45,33]
df = pd.DataFrame({'days':days,'temp':temp})
sns.lineplot(x='days',y='temp',data=df)
```
